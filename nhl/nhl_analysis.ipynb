```python
games = pd.read_csv('all_teams.csv')
```

```python
games.sort_values('gameId')
```

```python
games = games[games['situation'] == 'all']
```

```python
cols_with_x = [col for col in games.columns if 'x' in col]
```

```python
games = games.drop(columns = cols_with_x)
```

Create diff columns for all these games

```python
# Create a copy to preserve original data if needed
games_diff = games.copy()

# Loop over all columns ending in 'For' and find corresponding 'Against' columns
for col in games.columns:
    if col.endswith('For'):
        col_against = col.replace('For', 'Against')
        if col_against in games.columns:
            diff_col_name = col.replace('For', 'Diff')
            games_diff[diff_col_name] = games[col] - games[col_against]
```

```python
# Identify columns to drop
cols_to_drop = [col for col in games_diff.columns if col.endswith('For') or col.endswith('Against')]

# Drop them from the DataFrame
games_diff = games_diff.drop(columns=cols_to_drop)
```

```python
games_diff = games_diff.drop(columns=['season', 'position', 'situation', 'name', 'playerTeam'])
```

```python
games_diff = games_diff[games_diff['goalsDiff'] != 0]
```

```python
games_diff['home_team_win'] = (games_diff['goalsDiff'] > 0).astype(int)
```
